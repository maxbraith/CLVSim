```python
import pandas as pd
import re
```

```python
df = pd.read_csv("timestampsYear.csv", names=range(231))
df
```

```python
df.columns = df.iloc[0]
df = df[1:]
df.reset_index(drop=True, inplace=True)
df.head()
```

```python
for i in range(1, 228):
    print(i)
    df[f'timestep-{i}'] = df[f'timestep-{i}'].apply(
        lambda x: ' -'.join([
            str(pd.to_datetime(
                re.sub(r'(\d{1,2})(st|nd|rd|th)', r'\1', x.split(' -')[0]).strip(),
                errors='coerce'
            )),
            *x.split(' -')[1:]
        ]) if isinstance(x, str) else x
    )
```

```python
df.to_csv("timestampsDatetime.csv", index=False)
```
